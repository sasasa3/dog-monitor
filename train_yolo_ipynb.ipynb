```python
!pip install -U ultralytics

# --- è¿½åŠ ãƒ‡ãƒ¼ã‚¿ç”¨ï¼šAwakeç‰¹åŒ–ã‚¢ãƒãƒ†ãƒ¼ã‚·ãƒ§ãƒ³ ---
from ultralytics import YOLO
import os
import glob
from tqdm.notebook import tqdm

# --- è¨­å®š ---
# æ–°ã—ã„ç”»åƒã‚’å…¥ã‚ŒãŸãƒ•ã‚©ãƒ«ãƒ€
IMAGE_DIR = "/content/dataset_awake_only"
OUTPUT_DIR = "/content/dataset_awake_only/labels"

# â˜…é‡è¦â˜… ä»Šå›ã¯ã€Œèµ·ãã¦ã„ã‚‹çŠ¶æ…‹ã€ã‚’å¼·åŠ›ã«æ¤œå‡ºã™ã‚‹è¨€è‘‰ã ã‘ã‚’å…¥ã‚Œã‚‹
# ã‚¯ãƒ©ã‚¹IDã¯ã€Œ1ã€ã«å›ºå®šã—ã¾ã™
PROMPTS = ["dog walking", "dog standing", "dog sitting", "dog playing", "dog looking up"]
TARGET_CLASS_ID = 1  # 1 = awake

CONF_THRESHOLD = 0.15 # å°‘ã—æ„Ÿåº¦é«˜ã‚

# ------------------
os.makedirs(OUTPUT_DIR, exist_ok=True)
model = YOLO('yolov8x-worldv2.pt')

# è¤‡æ•°ã®ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’ã¾ã¨ã‚ã¦ã‚»ãƒƒãƒˆ
model.set_classes(PROMPTS)

image_paths = glob.glob(os.path.join(IMAGE_DIR, "*.jpg")) # jpgã®å ´åˆ
if not image_paths:
    image_paths = glob.glob(os.path.join(IMAGE_DIR, "*.png")) # pngã®å ´åˆ

print(f"å‡¦ç†é–‹å§‹: {len(image_paths)}æš (Awakeãƒ¢ãƒ¼ãƒ‰)")

for img_path in tqdm(image_paths):
    results = model.predict(img_path, conf=CONF_THRESHOLD, verbose=False)

    filename = os.path.basename(img_path)
    txt_name = os.path.splitext(filename)[0] + ".txt"
    save_path = os.path.join(OUTPUT_DIR, txt_name)

    content = []
    for result in results:
        for box in result.boxes:
            # ä½•ãŒè¦‹ã¤ã‹ã£ã¦ã‚‚ã€ä»Šå›ã¯å¼·åˆ¶çš„ã«ã€Œ1 (awake)ã€ã¨ã—ã¦ä¿å­˜ã™ã‚‹
            # (èµ·ãã¦ã„ã‚‹å‹•ç”»ã—ã‹å…¥ã‚Œã¦ã„ãªã„å‰æã®ä½œæˆ¦ã§ã™)
            xywh = box.xywhn[0].tolist()
            line = f"{TARGET_CLASS_ID} {xywh[0]:.6f} {xywh[1]:.6f} {xywh[2]:.6f} {xywh[3]:.6f}"
            content.append(line)

    with open(save_path, "w") as f:
        if content:
            f.write("\n".join(content))

print("âœ… Awakeãƒ‡ãƒ¼ã‚¿ã®ä½œæˆå®Œäº†ï¼ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„ã€‚")
```

```python
from ultralytics import YOLO
import glob
import random
import cv2
import matplotlib.pyplot as plt
import os

# 1. ãƒ¢ãƒ‡ãƒ«ã®èª­ã¿è¾¼ã¿ï¼ˆæœ€æ–°ã®å­¦ç¿’çµæœã‚’ä½¿ç”¨ï¼‰
runs_dir = '/content/runs/detect'
latest_train = max(glob.glob(os.path.join(runs_dir, 'train*')), key=os.path.getmtime)
best_model_path = os.path.join(latest_train, 'weights', 'best.pt')

print(f"ãƒ¢ãƒ‡ãƒ«èª­ã¿è¾¼ã¿ä¸­: {best_model_path}")
model = YOLO(best_model_path)

# 2. æ¤œè¨¼ç”¨ç”»åƒã®ãƒªã‚¹ãƒˆå–å¾—
# datasetãƒ•ã‚©ãƒ«ãƒ€ã®æ§‹æˆã«åˆã‚ã›ã¦æ‹¡å¼µå­ã‚’æŒ‡å®š (.png ã¾ãŸã¯ .jpg)
valid_images = glob.glob('/content/dataset/valid/images/*.*')

if not valid_images:
    print("ã‚¨ãƒ©ãƒ¼: æ¤œè¨¼ç”¨ç”»åƒãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚ãƒ‘ã‚¹ã‚’ç¢ºèªã—ã¦ãã ã•ã„ã€‚")
else:
    # 3. ãƒ©ãƒ³ãƒ€ãƒ ã«3æšé¸ã‚“ã§æ¨è«–ãƒ»è¡¨ç¤º
    print(f"æ¤œè¨¼ç”»åƒæ•°: {len(valid_images)}æšã‹ã‚‰ã€3æšã‚’ãƒ©ãƒ³ãƒ€ãƒ ãƒ†ã‚¹ãƒˆã—ã¾ã™...")

    # ç”»åƒã‚µã‚¤ã‚ºã‚’èª¿æ•´ã—ã¦è¡¨ç¤ºã™ã‚‹è¨­å®š
    plt.figure(figsize=(15, 5))

    picked_images = random.sample(valid_images, min(3, len(valid_images)))

    for i, img_path in enumerate(picked_images):
        # æ¨è«–å®Ÿè¡Œ
        results = model.predict(img_path, conf=0.5) # ç¢ºä¿¡åº¦0.5ä»¥ä¸Šã‚’è¡¨ç¤º

        # çµæœã®æç”»ï¼ˆYOLOã®æ©Ÿèƒ½ã§æ ä»˜ãç”»åƒã‚’ä½œæˆï¼‰
        result_array = results[0].plot()

        # Matplotlibã§è¡¨ç¤ºã™ã‚‹ãŸã‚ã«è‰²ã‚’å¤‰æ› (BGR -> RGB)
        result_rgb = cv2.cvtColor(result_array, cv2.COLOR_BGR2RGB)

        # æ¨ªã«ä¸¦ã¹ã¦è¡¨ç¤º
        plt.subplot(1, 3, i+1)
        plt.imshow(result_rgb)
        plt.axis('off')
        plt.title(f"Test {i+1}")

    plt.show()
```
